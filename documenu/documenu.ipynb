The API is queried to get restaurant information within a given zip code, for this example. [Documentation can be found here](https://documenu.com/docs#get_started).

## A function that can take a zipcode and return restaurant information from that area

```python
def restaurants_by_zip(zip_code):
    API_KEY = get_mp_api_key()
    url = f'https://api.documenu.com/v2/restaurants/zip_code/{zip_code}'
    payload = {
        # 'size': '5',
        'key': API_KEY,
        'fullmenu': 'false'
    }

    r = requests.get(url, params=payload)
    response = r.json()['data']
    df = pd.json_normalize(response)
    split_cuisines = pd.DataFrame(df.cuisines.tolist(), columns=['cuisine_1', 'cuisine_2', 'cuisine_3', 'cuisine_4', 'cuisine_5', 'cuisine_6'])
    df_final = pd.concat([df, split_cuisines], axis=1)
    df_final.drop(columns=['cuisines', 'menus'], inplace=True)

    return df_final
```
